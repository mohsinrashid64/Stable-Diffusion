```python
data=response.json()
```

```python
img=data['images'][0]
```

```python
import base64
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

# Let's say this is your base64 string (without the `data:image/png;base64,` prefix)
base64_str = img  # <-- full string here

# Decode base64
image_data = base64.b64decode(base64_str)

# Convert to image
image = Image.open(BytesIO(image_data))

# Plot the image
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```
